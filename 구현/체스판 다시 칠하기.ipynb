```python
# 2초이고 n이 50 이하이므로 완전탐색
n, m = map(int,input().split())
board = []
for i in range(n):
    board.append(list(input()))

# 완전탐색하기 위해 자르는 범위에 대한 변수 필요 : 시작점만 잡으면 될 듯  ex) 2~9칸 자르기
# 잘린 보드에서 잘못된 것 찾기 : 변을 공유하는 두 칸이 달라야 함 -> row이전과 같다면 현재 칸 수정 행 달라지면 윗 행 고려, 잘못된 개수 +1 
```

```python
# 첫번쨰 접근 : 움직이면서 해당 값이 주변 값과 다른지 비교
# -> 다음 값이 이전 값이 바뀌면서 제대로 바뀌지 않음
import copy

# print(b)
cnt = 1e5
for i in range(n - 7):  # 행 시작점, 보드 자르기
    for j in range(m - 7):  # 열 시작점
        b = copy.deepcopy(board)  # 시작할 때마 보드 초기화해줘야 함
        tmp_cnt = 0
        #         print(i,j)

        for dy in range(i, i + 8):  # 잘린 보드마다 확인
            for dx in range(j, j + 7):  # 첫번째 칸부터 확인 (dx-1과 비교하므로)
                if dx == j + 6 and b[dy][dx] == b[dy][dx + 1]:  # 끝칸 변경
                    tmp_cnt += 1
                    if b[dy][dx + 1] == 'W':
                        b[dy][dx + 1] = 'B'
                    else:
                        b[dy][dx + 1] = 'W'
                elif dy == 0 and b[dy][dx] == b[dy][dx + 1]:  # 첫 행에서는 옆이랑만 비교
                    tmp_cnt += 1
                    if b[dy][dx] == 'W':
                        b[dy][dx] = 'B'
                    else:
                        b[dy][dx] = 'W'
                elif dy > 0 and (b[dy][dx] == b[dy][dx + 1] or b[dy][dx]
                                 == b[dy - 1][dx]):  # 두 번째부터는 위, 옆 비교
                    tmp_cnt += 1
                    if b[dy][dx] == 'W':
                        b[dy][dx] = 'B'
                    else:
                        b[dy][dx] = 'W'
                print(b[dy][j:j + 8], tmp_cnt)
        if cnt > tmp_cnt:  # 최소 개수 갱신
            cnt = tmp_cnt
#         print(tmp_cnt)
print(cnt)
```

```python
# 참조 접근
# 첫 값을 W,B 둘 중 하나로 잡고 그에 맞춰 계산
# 첫 값이 W(0,0)이면 다른 W도 x,y 좌표 합이 짝수여야 함
# 한 번에 둘 다 계산 

cnt = 1e5

for i in range(n - 7):  # 행 시작점, 보드 자르기
    for j in range(m - 7):  # 열 시작점
        # 그냥 값 확인만 하므로 copy 필요 없음
        tmp_cnt_w = 0
        tmp_cnt_b = 0
        for dy in range(i, i + 8):
            for dx in range(j, j + 8):
                if (dx + dy) % 2 == 0:  # 첫 값 나와야 하는 부분
                    if board[dy][dx] == 'B':  # 첫 값이 W일때
                        tmp_cnt_w += 1
                    else:
                        tmp_cnt_b += 1
                else:  # 다른 값 나와야 하는 부분
                    if board[dy][dx] == 'W':
                        tmp_cnt_w += 1
                    else:
                        tmp_cnt_b += 1

#         print(tmp_cnt_w, tmp_cnt_b)
        if cnt > min(tmp_cnt_w, tmp_cnt_b):  # 최소 개수 갱신
            cnt = min(tmp_cnt_w, tmp_cnt_b)
#         print(tmp_cnt)
print(cnt)
```
