```python
def rotate(query, arr): # 회전 함수 
    tmp=[]
    for row in range(query[0]-1,query[2]): # 인덱스이므로 하나씩 빼줘야 함
        for column in range(query[1]-1,query[3]):
            if column == query[1]-1 and row == query[0]-1: #시작
                tmp.append(arr[row][column])
                arr[row][column] = arr[row+1][column]
            elif  row == query[0]-1: # 첫 행 -> column 회전하면서 값 작아짐
                tmp.append(arr[row][column])
                arr[row][column] = arr[row][column]-1 
             #  둘째에서 마지막 이전 행 테두리 -> 왼쪽은 row길이만큼 더하기
            elif column == query[1]-1 :
                tmp.append(arr[row][column])
                arr[row][column] = arr[row][column]+len(arr[0]) 
             #  둘째에서 마지막 이전 행 테두리 -> 오른쪽은 row길이만큼 빼기
            elif column == query[3]-1: # 행 밀기
                tmp.append(arr[row][column])
                arr[row][column] = arr[row][column]-len(arr[0])
            elif row == query[2]-1:
                tmp.append(arr[row][column])
                arr[row][column] = arr[row][column]+1      
#             print(arr)
#             print(tmp)
    return arr,min(tmp)
            
            

def solution(rows, columns, queries):
    arr = []
    answer = []
    for r in range(rows): # 일단 row, columns으로 행렬 만들기
        temp=[]
        for c in range(1,columns+1):
            temp.append(c+r*columns)
        arr.append(temp)  
    for query in queries:
        arr, min_value = rotate(query, arr)
        answer.append(min_value)
          
    return answer
```

```python
solution(6,6,[[2,2,5,4],[3,3,6,6],[5,1,6,3]])
```
