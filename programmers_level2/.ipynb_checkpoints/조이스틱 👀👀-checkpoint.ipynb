```python
# 내 접근 
# 여기서 만지다가 A - Z 이동, 커서 뒤로 이동하는 것까지 고려해야함을 깨닫고 풀이 참조

from string import ascii_uppercase
alpha_list = list(ascii_uppercase)

def solution(name):
    answer = 0
    start = 'A'*len(name)
    for i, alpha in enumerate(name): # 각 알파벳마다 확인
        moves = 10e5  # move 초기화
        for j in range(i,0,-1): # 처음(A)에서 거리, 이전 커서에서 거리, ..., 종합하여 min
            tmp_moves = abs(alpha_list.index(alpha) - alpha_list.index(name[j])) + j
            # 위치의 알파벳에서 목표까지 거리 + 커서 이동 거리
            if tmp_moves < moves: # 더 적으면 갱신
                moves = tmp_moves
        answer += moves 
            
    return answer
```

```python
# A-Z는 가중치 할당하는 방식 참조
# 커서 이동은 A를 제외한 모든 곳으로 이동하는 방식 참조

def solution(name):
    answer = 0
    n = len(name)

    def alphabet_to_num(char):
        num_char = [i for i in range(14)] + [j for j in range(12, 0, -1)] # A-Z 순환 고려하여 '알파벳마다 가중치 설정'
        return num_char[ord(char) - ord('A')] # 아스키 코드로 차이 계산 -> 그 위치의 가중치 할당

    for ch in name:
        answer += alphabet_to_num(ch)  # 왜 더하지..? 
        # 커서 이동이 다음에 입력할 '값'이 다른 커서위치의 '값'으로 변하는 것이 아니라
        # 내가 변경하고자 하는 위치로 이동만 하는 것.. / 문제를 잘못 이해했었음 
        # 따라서 각 위치에서 변경해야하는 것까지 이동 필요 -> 가중치 더하기

    move = n - 1
    for idx in range(n): # 커서 이동 빠른 곳으로(앞 or 뒤)
        next_idx = idx + 1
        while (next_idx < n) and (name[next_idx] == 'A'):   
            next_idx += 1
        distance = min(idx, n - next_idx) # 현재 인덱스, 뒤로 돌아가 다음 알파벳 만나는 경우 비교 
        # -> 시작에서 idx까지의 거리가 가까운지, 최종 위치까지가 가까운지 
        move = min(move, idx + n - next_idx + distance) # 위의 두 방향 가긴 다가야함 대신 커서 변경시 돌아와야 하므로(뒤로가면 다시 앞으로와 같이)
        # 두 방향 + 그 방향 중 가까운 거리 더함
        print('distance:',distance)
        print('move:',move)

    answer += move
    return answer
```

```python
# product 라이브러리 사용 -> brute force 방법으로 시간 많이 소요
from collections import deque
from itertools import product
# itertools.product()는 두개 이상의 리스트(or 집합) 끼리의 데카르트 곱(cartesian product)를 계산하여 iterator로 반환해준다.
# ((x,y) for x in A for y in B) == 중복조합
# repeat은 튜플의 크기
#A = [1,2,3]
#list(product(A, repeat=2))
# >>>> [(1, 1), (1, 2), (1, 3), (2, 1), (2, 2), (2, 3), (3, 1), (3, 2), (3, 3)]

def solution(name):    
    results = []

    for rs in product((-1,1), repeat=len(name)-1): # 1, -1 이동으로 모든 경우의 수 따짐
        # 이렇게 하면 이중에 아예 최대 길이안에 되지 않는 경우도 있어 비효율적이긴함(len9의 경우 256 경우 중 78개만) ->Brute force
        name_deque = deque(name)
        default = deque('A'*len(name))
        
        for c, r in enumerate([0]+list(rs)): # 최대 이동횟수는 list길이 이하 이므로 c로 충분 
            #print(c,r)
            default.rotate(r) 
#           deq = deque([1, 2, 3, 4, 5])
#           deq.rotate(1)
#           print(deq) ### [5, 1, 2, 3, 4] # 한번 밀렸다가 
#           deq.rotate(-1) ### [1, 2, 3, 4, 5] # 다시 돌아옴 
            name_deque.rotate(r)
            #print(default,name_deque)
            default[0] = name_deque[0] # default가 모두 A므로 값 가져와야 함 

            if name_deque == default:
                results.append(c) # 이동 횟수 
                break
    print(len(results))
        
    return min(set(results))+sum([min(ord(l)-ord('A'), ord('Z')-ord(l)+1) for l in name]) # 이동횟수 + 알파벳 변경횟수
```

```python
def solution(name):
    init_A = list("A"*len(name))
    name=list(name)
    TF=[]
    for a,b in zip(name,init_A):
        TF.append(a==b)  
    count=move_count(TF,0,0) # 이동횟수
    for alpa in name:
        tmp=ord(alpa)-ord("A") # 없어도 됨
        count+=min(ord(alpa)-ord("A"),ord("Z")-ord(alpa)+1) # 알파벳 변경
    return count

def move_count(TF,point,count):
    if sum(TF)==len(TF): # sum(TF) == 맞는 부분 
        return count
    else:
        tmp_all=[]
        target_index=[]
        tmp_TF=TF.copy()
        target_index_count=0
        
        while len(target_index) <= 1:
            if not TF[(point+target_index_count)%len(TF)]:
                target_index.append((point+target_index_count)%len(TF))
            if not TF[(point-target_index_count)%len(TF)]:
                target_index.append((point-target_index_count)%len(TF))
            target_index_count+=1
        #print("TF:",TF)
        #print("point:",point)
        #print("접근:",set(target_index))

        for i in set(target_index):
            if not TF[i]:
                tmp_TF=TF.copy()
                tmp_TF[i]=True
                tmp_all.append(move_count(tmp_TF,i,count+min(len(TF)-abs(point-i),abs(point-i)))) # 재귀
        return min(tmp_all)
```

```python
# 맨 위 풀이와 유사하나 조금더 직관적
def solution(name):

	# 조이스틱 조작 횟수 
    answer = 0
    
    # 기본 최소 좌우이동 횟수는 길이 - 1
    min_move = len(name) - 1
    
    for i, char in enumerate(name):
    	# 해당 알파벳 변경 최솟값 추가
        answer += min(ord(char) - ord('A'), ord('Z') - ord(char) + 1)
        
        # 해당 알파벳 다음부터 연속된 A 문자열 찾기
        next = i + 1
        while next < len(name) and name[next] == 'A':
            next += 1
            
        # 기존, 연속된 A의 왼쪽시작 방식, 연속된 A의 오른쪽시작 방식 비교 및 갱신
        min_move = min([min_move, 2 *i + len(name) - next, i + 2 * (len(name) -next)])
        
    # 알파벳 변경(상하이동) 횟수에 좌우이동 횟수 추가
    answer += min_move
    return answer
```
