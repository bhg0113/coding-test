```python
def flip(string):
    for i,v in string:
        if string[i]=='(':
            string[i] = ')'
        else:
            string[i] = '('
    return string
def correct(string): # 올바른 문자열인지 확인
    for i, v in enumerate(string):
        if string[:i].count('(') < string[:i].count(')'): 
            return False
        else:
            return True 
def divide_uv(p): # 균형잡힌 문자열을 구분으로 u,v로 나눔
    for i,v  in enumerate(p):
        if p[:i].count('(') == p[:i].count(')'): 
            print(i)
            u,v = p[:i], p[i:] # 여기서 인덱싱이 제대로 안됨 -> 왜? 
    return u,v 

def solution(p):
    u,v = divide_uv(p)
    print(u,v)
    if correct(u): # 올바른 문자열인지 확인 
        while v != '':
            solution(v)
        return u+v
    else:
        return '(' + solution(v)[1] + ')' + flip(u[1:-1])
```
