```python
# 내 접근 
# 여기서 만지다가 A - Z 이동, 커서 뒤로 이동하는 것까지 고려해야함을 깨닫고 풀이 참조

from string import ascii_uppercase
alpha_list = list(ascii_uppercase)

def solution(name):
    answer = 0
    start = 'A'*len(name)
    for i, alpha in enumerate(name): # 각 알파벳마다 확인
        moves = 10e5  # move 초기화
        for j in range(i,0,-1): # 처음(A)에서 거리, 이전 커서에서 거리, ..., 종합하여 min
            tmp_moves = abs(alpha_list.index(alpha) - alpha_list.index(name[j])) + j
            # 위치의 알파벳에서 목표까지 거리 + 커서 이동 거리
            if tmp_moves < moves: # 더 적으면 갱신
                moves = tmp_moves
        answer += moves 
            
    return answer
```

```python
# A-Z는 가중치 할당하는 방식 참조
# 커서 이동은 A를 제외한 모든 곳으로 이동하는 방식 참조

def solution(name):
    answer = 0
    n = len(name)

    def alphabet_to_num(char):
        num_char = [i for i in range(14)] + [j for j in range(12, 0, -1)] # A-Z 순환 고려하여 '알파벳마다 가중치 설정'
        return num_char[ord(char) - ord('A')] # 아스키 코드로 차이 계산 -> 그 위치의 가중치 할당

    for ch in name:
        answer += alphabet_to_num(ch)  # 왜 더하지..? 
        # 커서 이동이 다음에 입력할 '값'이 다른 커서위치의 '값'으로 변하는 것이 아니라
        # 내가 변경하고자 하는 위치로 이동만 하는 것.. / 문제를 잘못 이해했었음 
        # 따라서 각 위치에서 변경해야하는 것까지 이동 필요 -> 가중치 더하기

    move = n - 1
    for idx in range(n): # 커서 이동 빠른 곳으로(앞 or 뒤)
        next_idx = idx + 1
        while (next_idx < n) and (name[next_idx] == 'A'):   
            next_idx += 1
        distance = min(idx, n - next_idx) # 현재 인덱스, 뒤로 돌아가 다음 알파벳 만나는 경우 비교 
        # -> 시작에서 idx까지의 거리가 가까운지, 최종 위치까지가 가까운지 
        move = min(move, idx + n - next_idx + distance) # 위의 두 방향 가긴 다가야함 대신 커서 변경시 돌아와야 하므로(뒤로가면 다시 앞으로와 같이)
        # 두 방향 + 그 방향 중 가까운 거리 더함
        print('distance:',distance)
        print('move:',move)

    answer += move
    return answer
```

```python
# product 라이브러리 사용 -> brute force 방법으로 시간 많이 소요
from collections import deque
from itertools import product
# itertools.product()는 두개 이상의 리스트(or 집합) 끼리의 데카르트 곱(cartesian product)를 계산하여 iterator로 반환해준다.
# ((x,y) for x in A for y in B) == 중복조합
# repeat은 튜플의 크기
#A = [1,2,3]
#list(product(A, repeat=2))
# >>>> [(1, 1), (1, 2), (1, 3), (2, 1), (2, 2), (2, 3), (3, 1), (3, 2), (3, 3)]

def solution(name):    
    results = []

    for rs in product((-1,1), repeat=len(name)-1):
        name_deque = deque(name)
        default = deque('A'*len(name))
        #print(list(rs))

        for c, r in enumerate([0]+list(rs)):
            print(c,r)
            default.rotate(r) 
#           deq = deque([1, 2, 3, 4, 5])
#           deq.rotate(1)
#           print(deq) ### [5, 1, 2, 3, 4] # 한번 밀렸다가 
#           deq.rotate(-1) ### [1, 2, 3, 4, 5] # 다시 돌아옴 
            name_deque.rotate(r)
            default[0] = name_deque[0] #

            if name_deque == default:
                results.append(c)
                break

    return min(set(results))+sum([min(ord(l)-ord('A'), ord('Z')-ord(l)+1) for l in name])
```
