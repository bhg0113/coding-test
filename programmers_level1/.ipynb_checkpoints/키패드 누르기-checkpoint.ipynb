```python
def solution(numbers, hand):
    # 이중 리스트로 다이얼을 만들어 두고
    # 각 칸의 위아래 차이로 거리 계산하여 출력 -> 인덱스 찾을 수 없음 -> 반복문으로 구하기?
    dial_left = [1,4,7]
    dial_right= [3,6,9]
    dial_middle = [2,5,8,0]
    answer = ''
    left, right, h, tmp = 4,4,0,0
    lm, rm = False, False # 중간인 경우 거리 -1 하기 위해 어쩔 수 없이 도입
    for n in numbers:
        if n in dial_left:
            left = dial_left.index(n)+1
            lm = False
            h = 'L'
        elif n in dial_right:
            right = dial_right.index(n)+1
            rm = False 
            h = 'R'
        else:
            tmp = dial_middle.index(n)+1 
            if lm:
                l = abs(left - tmp)-1 
            else:
                l = abs(left - tmp)
            if rm:
                r = abs(right - tmp)-1
            else:
                r = abs(right - tmp)
            if l==r:
                if hand == 'left': 
                    left = tmp 
                    lm = True
                    h = 'L'
                else:
                    right =tmp
                    rm = True
                    h = 'R'
            else:
                if l<r:
                    left = tmp
                    lm= True
                    h = 'L'
                else:
                    right =tmp
                    rm = True
                    h = 'R'
        
        answer += h
    return answer
```

```python
# 처음 생각했던 딕셔너리 이용한 방법
def solution(numbers, hand):
    answer = ''
    key_dict = {1:(0,0),2:(0,1),3:(0,2),
                4:(1,0),5:(1,1),6:(1,2),
                7:(2,0),8:(2,1),9:(2,2),
                '*':(3,0),0:(3,1),'#':(3,2)}

    left = [1,4,7]
    right = [3,6,9]
    lhand = '*'
    rhand = '#'
    for i in numbers:
        if i in left:
            answer += 'L'
            lhand = i
        elif i in right:
            answer += 'R'
            rhand = i
        else:
            curPos = key_dict[i]
            lPos = key_dict[lhand]
            rPos = key_dict[rhand]
            ldist = abs(curPos[0]-lPos[0]) + abs(curPos[1]-lPos[1])
            rdist = abs(curPos[0]-rPos[0]) + abs(curPos[1]-rPos[1])

            if ldist < rdist:
                answer += 'L'
                lhand = i
            elif ldist > rdist:
                answer += 'R'
                rhand = i
            else:
                if hand == 'left':
                    answer += 'L'
                    lhand = i
                else:
                    answer += 'R'
                    rhand = i

    return answer
```
